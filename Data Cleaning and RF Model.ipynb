```python
# I am going to drop cloud 9am, cloud 3pm, evaporation, and sunshine because these have a high percentage of NA values
data.drop(['Cloud9am', 'Cloud3pm', 'Evaporation', 'Sunshine'], axis = 1, inplace = True)
```

```python
data.columns
```

```python
# Going to engineer some new columns based on notes with the team
# Making seasons
def get_season(month):
    if month in [1,2,12]: return 'Summer'
    elif month in [3,4,5]: return 'Fall'
    elif month in [6,7,8]: return 'Winter'
    elif month in [9,10,11]: return 'Spring'

data['Month'] = data['Date'].dt.month
data['Season'] = data['Month'].apply(lambda x: get_season(x))
# Change in temperature
data['Temperature_Change'] = data['MaxTemp'] - data['MinTemp']
# Average daily pressure
data['Average_Pressure'] = (data['Pressure9am'] + data['Pressure3pm']) / 2
# Average humidity
data['Average_Pressure'] = (data['Humidity9am'] + data['Humidity3pm']) / 2
# Average wind speed
data['Average_Windspeed'] = (data['WindSpeed9am'] + data['WindSpeed3pm']) / 2
print(data.Season.value_counts(normalize = True))
```

```python
# making lists of number and category columns
list_num = data.dtypes != 'object'
list_obj = data.dtypes == 'object'
# indexing the data frame using the column lists
num_col = data.columns[list_num].tolist()
cat_col = data.columns[list_obj].tolist()
```

```python
#### SIMPLE IMPUTER #### To fill the NAs
from sklearn.impute import SimpleImputer

imputer_num_mean = SimpleImputer(strategy = 'mean')
imputer_category = SimpleImputer(strategy = 'most_frequent')

```

```python
# For loop to apply imputation
for col in data.columns.tolist():
    if col in num_col:
        data[col] = imputer_num_mean.fit_transform(data[[col]])
    else:
        data[col] = imputer_category.fit_transform(data[[col]])
```
